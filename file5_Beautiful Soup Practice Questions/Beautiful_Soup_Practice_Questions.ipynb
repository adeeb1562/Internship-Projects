```python
headr.text
```

```python
#cuisine
headr=soup.find('div', class_="detail-info")
headr
```

```python
headr.text
pattern = r'\|(.*)'
match = re.search(pattern, headr.text)
cuisine = match.group(1).strip() if match else ""
print(cuisine)
```

# 4)	Write s python program to display list of respected former finance minister of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
pg=requests.get('https://presidentofindia.nic.in/former-presidents')
pg
```

```python
soup=BeautifulSoup(pg.content)
soup
```

```python
name=soup.find_all('div', class_="desc-sec")
name
```

```python
names = []
terms_of_office = []

for div in name:
    h3_content = div.find('h3').text.strip()
    h5_content = div.find('h5').text.strip()

    names.append(h3_content)
    terms_of_office.append(h5_content)

df = pd.DataFrame({'Name': names, 'Term of Office': terms_of_office})
print(df)
```
